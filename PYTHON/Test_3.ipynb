1.Explain the difference between lists and tuples in Python. Why might you choose one over the other?

In Python, both lists and tuples are useful for storing collections of items, but they serve different purposes. Lists are like flexible containers,you can change,add,or remove items in them as needed. This flexibility comes from the fact that lists are mutable. They use square brackets, like [1,2,3], making them perfect for situations where your data might need updates, such as a list of tasks or user inputs.

Tuples, on the other hand, are more like locked containers. Once you create them, you can't change their content because they are immutable. They use parentheses, like (1,2,3), and this fixed nature makes them more memory-efficient and faster than lists. Tuples are great when you want to ensure that your data stays constant, such as coordinates, settings, or data that should be protected from accidental changes.
Deciding whether to use a list or a tuple depends on what you need: choose a list if flexibility is important, and go with a tuple when performance and immutability are your priority.

2.Write a function that takes a list of integers and returns a new list with only the even numbers.

```python
def get_even_numbers(numbers):
    even_list = []
    for num in numbers:
        if num % 2 == 0:
            even_list.append(num)
    return even_list
input_list = []
print("Enter integers one by one. Type 'done' when you are finished:")
while True:
    user_input = input()
    if user_input.lower() == 'done':
        break
    try:
        input_list.append(int(user_input))
    except ValueError:
        print("Please enter a valid integer or 'done' to finish.")
even_numbers = get_even_numbers(input_list)
print("Even numbers:", even_numbers)
```

3.Describe how Python's garbage collection works. What are the main methods it uses to manage memory?

Python's garbage collection is a smart feature that helps manage memory automatically, making sure that unused or unreachable objects don't take up space unnecessarily.

Python uses a method called reference counting to manage memory by keeping track of how many references point to each object. Each object has a counter that increases when it gets a new reference and decreases when a reference is deleted. When the counter hits zero, it means no one is using that object anymore, so Python can free up its memory. For instance, if you create a list and assign it to a new variable, the count goes up. But when you delete both variables, the count drops to zero, and the list is cleared from memory.

While reference counting is effective, it struggles with cyclic references, where two or more objects reference each other, creating a loop that keeps their counts from dropping to zero. To solve this, Python has a built-in garbage collector that identifies and cleans up these cycles.


4.Given a dictionary where keys are student names and values are lists of their scores, write a function that returns the average score for each student.

```python
def calculate_average_scores(students_scores):
    average_scores = {}
    for student, scores in students_scores.items():
        if scores:
            average_scores[student] = sum(scores) / len(scores)
        else:
            average_scores[student] = 0
    return average_scores
students_scores = {}
num_students = int(input("How many students do you want to enter? "))
for i in range(num_students):
    name = input("What is the student's name? ")
    scores_input = input(f"Please enter the scores for {name}, separated by commas: ")
    scores = []
    for score in scores_input.split(','):
        scores.append(int(score))
    students_scores[name] = scores
average_scores = calculate_average_scores(students_scores)
print("Here are the average scores for each student:")
print(average_scores)
```

5.What are *args and **kwargs in Python functions, and how are they used?

*args is used for non-keyword variable-length arguments and is treated as a tuple.
**kwargs is used for keyword variable-length arguments and is treated as a dictionary. These features provide flexibility in function calls, allowing you to write more dynamic and adaptable functions.







```python
def args(*args):
    for arg in args:
        print(arg)
args(11,23,56,'Aravind')

def kwargs(**kwargs):
    for i, j in kwargs.items():
        print(f"{i}: {j}")
kwargs(name='Aravind', age=26, address='Kannur')
```

1.Explain how Streamlit's st.cache works and why it is useful. When should you avoid using it?

st.cache is a useful feature that speeds up your applications by storing the results of time-consuming functions. When you use @st.cache on a function, Streamlit remembers the result the first time you run it with specific inputs. So, if you call it again with the same inputs, it simply uses the stored result instead of recalculating everything.

Caching is great because it makes your app faster and more responsive, especially when working with large data sets. It ensures that you get consistent results when the inputs dont change, so you dont have to keep repeating heavy calculations. Streamlit also automatically clears the cache if your code or inputs change, so you can rely on having the latest data.

However, there are times when you should avoid using caching. If your function changes something outside of itself, like saving to a file, caching might cause confusion since it wont re-run the function when those changes happen. Caching can also use a lot of memory, especially with large datasets, which could slow down your app. Lastly, if your inputs change frequently, caching wont help much since the cache keeps getting cleared, leading to constant recalculations. In these situations, it's important to think about whether caching is really the right choice for your app

4.Create a Streamlit app that displays a line chart of random data. Allow the user to select the number of data points they want in the chart using a slider.

```python
import streamlit as st
import numpy as np
import pandas as pd
st.title("Line Chart")
points = st.slider("Select the number of data points:", min_value=1, max_value=10, value=1)
data = np.random.randn(points)
df = pd.DataFrame(data, columns=['Random Data'])
st.line_chart(df)
```

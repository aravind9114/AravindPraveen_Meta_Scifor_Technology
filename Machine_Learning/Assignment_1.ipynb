```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"Automobile_data.csv")
pd.set_option('display.max_columns',50)
df
```

```python
df.info()
```

```python
df = df.replace('?',np.NaN)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.dtypes
```

```python
columns_float = ['normalized-losses','bore','stroke','horsepower','peak-rpm','price']
df[columns_float] = df[columns_float].astype(float)
```

```python
columns_with_null=[]

for i in df:
    if df[i].isnull().sum()>0:
        columns_with_null.append(i)
columns_with_null
```

```python
mean_colms = ['normalized-losses','bore','stroke','horsepower','peak-rpm']

for i in mean_colms:
    df[i]=df[i].fillna(round(df[i].mean(),2))
```

```python
df['num-of-doors'] = df['num-of-doors'].fillna(df['num-of-doors'].mode()[0])
```

```python
df.dropna(subset=["price"], axis=0, inplace=True)
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
colms = ['normalized-losses','horsepower','peak-rpm']

df[colms] = df[colms].astype(int)
```

**What's the use of pandas, numpy, seaborn, matplotlib? Explain with example**

Pandas is a powerful library for data manipulation and analysis, especially with tabular data. It provides easy-to-use data structures such as DataFrame and Series.
Loading and handling datasets.
Data cleaning, filtering, and transformation.
Grouping and aggregating data for analysis.

```python
import pandas as pd
data = {'Name': ['Alex', 'Aravind', 'Praveen'],
        'Age': [25, 30, 35],
        'City': ['Delhi', 'Kerala', 'Mumbai']}
df = pd.DataFrame(data)
print(df)

```

NumPy is a fundamental package for scientific computing. It provides support for arrays, matrices, and many mathematical functions to perform operations on these arrays.
Working with arrays.
Fast mathematical computations.
Linear algebra, random number generation.

```python
import numpy as np
array = np.array([[1, 2, 3], [4, 5, 6]])
array
```

Seaborn is a statistical data visualization library built on top of Matplotlib. It makes it easy to create visually appealing and informative statistical plots.
Visualization of complex datasets with less code.
Plotting distribution, relationships, and categorical data.
Enhancing Matplotlib plots.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
data = {
    'Student': ['John', 'Emma', 'Sam', 'Olivia', 'Alex', 'Sophia'],
    'Math': [85, 90, 78, 92, 88, 75]
}

df = pd.DataFrame(data)
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='Student', y='Math', color='skyblue')
plt.title('Math Marks of Students', fontsize=14)
plt.xlabel('Student', fontsize=12)
plt.ylabel('Math Marks', fontsize=12)
plt.show()
```

Matplotlib is a widely used library for creating static, interactive, and animated visualizations in Python. It provides extensive controls for creating a variety of charts.
Creating plots such as line charts, bar charts, histograms.
Customizing figures.
Building simple or complex visualizations.

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.plot(x, y, label="Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Line Plot")
plt.show()
```

1.Descriptive Analysis

Visualization Example:
For the monthly sales data [120, 150, 180, 200, 250], a box plot can be used to visualize the data distribution.



```python
import matplotlib.pyplot as plt

sales = [120, 150, 180, 200, 250]
plt.boxplot(sales, vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title("Box Plot of Monthly Sales")
plt.xlabel("Units Sold")
plt.show()
```

2. Inferential Statistics


Does a new marketing strategy increase sales?

Sales before strategy: [200, 210, 215, 220, 230]

Sales after strategy: [250, 260, 270, 280, 300]

Python Implementation for T-Test:

```python
from scipy.stats import ttest_ind

before_sales = [200, 210, 215, 220, 230]
after_sales = [250, 260, 270, 280, 300]

t_stat, p_value = ttest_ind(before_sales, after_sales)
print(f"t-statistic: {t_stat}, p-value: {p_value}")

#Interpretation: If p < 0.05, the difference is statistically significant.
```

4. Frequency Distribution

```python
import matplotlib.pyplot as plt

scores = [70, 80, 90, 70, 80, 90, 90, 100, 70]
plt.hist(scores, bins=[70, 80, 90, 100, 110], edgecolor='black', color='lightgreen')
plt.title("Histogram of Exam Scores")
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.show()
```
